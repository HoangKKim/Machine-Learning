```python
import numpy as np

# Hàm mất mát
def loss_function(x, y):
    return x ** 2 + y ** 2

# Gradient của hàm mất mát
def gradient(x, y):
    grad_x = 2 * x
    grad_y = 2 * y
    return np.array([grad_x, grad_y])

# Gradient Descent
def gradient_descent(starting_point, learning_rate, epochs):
    point = np.array(starting_point)
    history = []  # Lưu trữ lịch sử các giá trị của (x, y)

    for i in range(epochs):
        grad = gradient(point[0], point[1])
        point = point - learning_rate * grad
        history.append(point.copy())
        print(f"Epoch {i+1}: x = {point[0]}, y = {point[1]}, f(x, y) = {loss_function(point[0], point[1])}")

    return point, history

# Khởi tạo
starting_point = [10, 10]  # Điểm khởi đầu
learning_rate = 0.1  # Bước nhảy
epochs = 50  # Số lần lặp

# Chạy thuật toán Gradient Descent
optimal_point, history = gradient_descent(starting_point, learning_rate, epochs)
print(f"\nGiá trị tối ưu của (x, y) là: {optimal_point}")
```
